```python
from pathlib import Path
import pandas as pd
```

```python
data_path = Path('LendingClubDataset')/ 'lending_club_loan_two.csv'
data_path1 = Path('LendingClubDataset')/ 'lending_club_info.csv'

df1 = pd.read_csv(data_path1)
df = pd.read_csv(data_path)
```

```python
cols_to_drop = ["emp_title","title","address"]
df= df.drop(columns=cols_to_drop)
```

```python
df.head().T
```

```python
df['loan_status'] = df['loan_status'].replace({"Fully Paid" :0,"Charged Off":1})
df['term']= df['term'].str.replace(" months","").astype(int)
```

```python
length_change = {"< 1 year":0,
                 "2 years":2,
                 "3 years":3,
                 "4 years":4,
                 "5 years":5,
                 "6 years":6,
                 "7 years":7,
                 "8 years":8,
                 "9 years":9,
                 "10+ years":10 }
df['emp_length'] = df['emp_length'].replace(length_change)
```
